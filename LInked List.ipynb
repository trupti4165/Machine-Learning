```python
#Q1.Reverse a Linked List:Write a Python function to reverse a singly linked list.

class Node:

    def __init__(self, data):
	    self.data = data
	    self.next = None


class LinkedList:

	# head function
	def __init__(self):
		self.head = None

	# To reverse the list
	def reverse(self):
		prev = None
		current = self.head
		while(current is not None):
			next = current.next
			current.next = prev
			prev = current
			current = next
		self.head = prev


	def push(self, new_data):
		new_node = Node(new_data)
		new_node.next = self.head
		self.head = new_node

	# To print the list
	def printL(self):
		temp = self.head
		while(temp):
			print(temp.data)
			temp = temp.next


# Main function
llist = LinkedList()
llist.push(1)
llist.push(2)
llist.push(3)
llist.push(4)

print("Given Linked List")
llist.printL()
llist.reverse()
print("\nReversed Linked List")
llist.printL()
```

```python
#Q5. Check for Balanced Parentheses:Write a Python function to check if a given string containing parentheses is balanced (i.e., every opening parenthesis has a corresponding closing parenthesis).
# Function to check if the parentheses in a string are balanced
def check(str):
	s = []
	for c in str:
		if c == '(':
			s.append('(')
		elif c == ')':
			if len(s) == 0:
				return 0
			else:
				p = s[-1]
				if p == '(':
					s.pop()
				else:
					return 0
	if len(s) == 0:
		return 1
	else:
		return 0


str = "()(())()()()"
if check(str) == 0:
	print("Not Balanced")
else:
	print("Balanced")
```

```python
#Q6. Find the Missing Number in an Array:Given an array containing n distinct numbers taken from 0, 1, 2, ..., n, write a Python function to find the one missing number.
def findMissingNumb(given_list):
    n = len(given_list)
    length = n + 1

    # Getting sum of all numbers from 1 to length using mathematical formula
    totalSum = length * (length + 1) / 2
    # calculating the sum of all elements in the given list
    arraysum = 0
    # Traversing the list
    for ele in given_list:
        arraysum = arraysum+ele
        missnum = totalSum-arraysum
    # reeturn the missing number
    return int(missnum)

# given list
given_list = [1, 3, 2, 6, 5]
print("The missing number in the given list", findMissingNumb(given_list))
```

```python
#Q8.Calculate Fibonacci Sequence:Write a Python function to calculate the nth Fibonacci number using both recursive and iterative approaches.
# Function for nth Fibonacci number using recursive
def Fibonacci(n):
	if n<= 0:
		print("Incorrect input")
	# First Fibonacci number is 0
	elif n == 1:
		return 0
	# Second Fibonacci number is 1
	elif n == 2:
		return 1
	else:
		return Fibonacci(n-1)+Fibonacci(n-2)

# Driver Program

print(Fibonacci(10))
```

```python
#Q9.Longest Common Subsequence Given two strings:write a Python function to find the length of their longest common subsequence.

def lcs(X, Y, m, n):

	if m == 0 or n == 0:
	    return 0;
	elif X[m-1] == Y[n-1]:
	    return 1 + lcs(X, Y, m-1, n-1);
	else:
	    return max(lcs(X, Y, m, n-1), lcs(X, Y, m-1, n));


# Driver program to test the above function
X = "AGGTAB"
Y = "GXTXAYB"
print ("Length of LCS is:", lcs(X, Y, len(X), len(Y)))
```
