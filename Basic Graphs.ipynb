```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[3,4,5,6,6]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'bo')
```

```python
plt.plot(y,'ro')
```

```python
plt.plot(y,'b+')
```

```python
plt.plot(x,y,'go--')
```

```python
plt.plot(x,y,'go--',linewidth='4',markersize='4')
```

```python
plt.bar(x,y,align='center',color=['b','g','y','b','r'])
```

```python
plt.barh(x,y,align='edge',color=['b','g','y','b','r'])
```
